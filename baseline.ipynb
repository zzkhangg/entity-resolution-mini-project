```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer(
                                lowercase=True,
                                stop_words="english",
                                analyzer='char_wb',
                                ngram_range=(2,3)
                            )
```

```python
def calculate_row_similarity(row):
    # Combine the two strings into a list for the vectorizer
    texts = [row['amazon_name'], row['google_name']]
    
    # If one is empty, similarity is 0
    if not texts[0] or not texts[1]:
        return 0.0
    
    # Transform the names into TF-IDF vectors
    tfidf_matrix = vectorizer.fit_transform(texts)
    
    # Calculate similarity between row 0 and row 1
    similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
    return similarity[0][0]
```
